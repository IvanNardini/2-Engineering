# Text Analysis for Women's E-Commerce Clothing Reviews

## Libraries and Settings

```python
import os
import tqdm
import pandas as pd

import string
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import PorterStemmer

%config Completer.use_jedi = False
```

## Variables

```python
INTERIM_DIR = os.path.join(os.pardir, 'data', 'interim')
RANDOM_STATE = 8
LANG = 'english'
```

## Helpers

```python
def remove_sw(words_list):
    stop_words=stopwords.words("english")
    return [word for word in words_list if word not in stop_words]

def stemmer(words_list):
    ps = PorterStemmer()
    return [ps.stem(word) for word in words_list]
```

## Load data

```python
X_train = pd.read_csv(os.path.join(INTERIM_DIR, 'X_train.csv'))
X_test = pd.read_csv(os.path.join(INTERIM_DIR, 'X_test.csv'))
X_val = pd.read_csv(os.path.join(INTERIM_DIR, 'X_val.csv'))
```

## Data Preparation

```python
# lower
X_train['review_lower'] = X_train['review_text'].apply(lambda x: x.lower())
# remove punctualizations
X_train['review_nopct'] = X_train['review_lower'].apply(lambda x: x.translate(str.maketrans('','',string.punctuation)))
# remove digit
X_train['review_nodg'] = X_train['review_nopct'].apply(lambda x: x.translate(str.maketrans('','',string.digits)))
# split 
X_train['review_word_tokens'] = X_train['review_nodg'].apply(word_tokenize)
# remove stopwords
X_train['review_no_sw'] = X_train['review_word_tokens'].apply(remove_sw)
# stemming
X_train['review_stem'] = X_train['review_no_sw'].apply(stemmer)
```

```python
X_test['review_lower'] = X_test['review_text'].apply(lambda x: x.lower())
X_test['review_nopct'] = X_test['review_lower'].apply(lambda x: x.translate(str.maketrans('','',string.punctuation)))
X_test['review_nodg'] = X_test['review_nopct'].apply(lambda x: x.translate(str.maketrans('','',string.digits)))
X_test['review_word_tokens'] = X_test['review_nodg'].apply(word_tokenize)
X_test['review_no_sw'] = X_test['review_word_tokens'].apply(remove_sw)
X_test['review_stem'] = X_test['review_no_sw'].apply(stemmer)
```

```python
X_val['review_lower'] = X_val['review_text'].apply(lambda x: x.lower())
X_val['review_nopct'] = X_val['review_lower'].apply(lambda x: x.translate(str.maketrans('','',string.punctuation)))
X_val['review_nodg'] = X_val['review_nopct'].apply(lambda x: x.translate(str.maketrans('','',string.digits)))
X_val['review_word_tokens'] = X_val['review_nodg'].apply(word_tokenize)
X_val['review_no_sw'] = X_val['review_word_tokens'].apply(remove_sw)
X_val['review_stem'] = X_val['review_no_sw'].apply(stemmer)
```
